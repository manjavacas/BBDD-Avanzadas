```
!pip install dnspython
import pymongo as pm

client = pm.MongoClient("mongodb+srv://mms:mms123@bdadatawarehouse-qalqb.mongodb.net/test?retryWrites=true&w=majority")
db = client["magallanes_nokia"]
```

Obtención datos NOKIA

```
year2017 = db["2017"]
for document in year2017.find({"description": "ACCIONES|NOKIA OYJ"}):
  eur_nokia_2017 = document["current_euros"]
  prct_nokia_2017 = document["current_percent"]

year2018 = db["2018"]
for document in year2018.find({"description": "ACCIONES|NOKIA OYJ"}):
  eur_nokia_2018 = document["current_euros"]
  prct_nokia_2018 = document["current_percent"]

year2019 = db["2019"]
for document in year2019.find({"description": "ACCIONES|NOKIA OYJ"}):
  eur_nokia_2019 = document["current_euros"]
  prct_nokia_2019 = document["current_percent"]
```

Gráficos NOKIA

```
import matplotlib.pyplot as plt

dataEurNokia = {'2017': eur_nokia_2017, '2018': eur_nokia_2018, '2019': eur_nokia_2019}
dataPercentNokia = {'2017': prct_nokia_2017, '2018': prct_nokia_2018, '2019': prct_nokia_2019}

namesEur = list(dataEurNokia.keys())
valuesEurNokia = list(dataEurNokia.values())
namesPrct = list(dataPercentNokia.keys())
valuesPrctNokia = list(dataPercentNokia.values())

fig, axs = plt.subplots(1, 1, figsize=(14,10))
axs.bar(namesEur, valuesEurNokia)
fig.suptitle('Nokia OYJ Actions (EUR)')

fig1, axs1 = plt.subplots(1, 1, figsize=(14,10))
axs1.bar(namesPrct, valuesPrctNokia)
fig1.suptitle('Nokia OYJ Actions (PERCENT)')
plt.show()
```
